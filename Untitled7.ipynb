## Extract the dataset

### Subtask:
Unzip the provided archive to access the image data.


**Reasoning**:
Unzip the provided archive to access the image data.



```python
import zipfile
import os

zip_file_path = 'archive (2).zip' # Corrected filename
extract_dir = '.'

if not os.path.exists(zip_file_path):
    print(f"Error: The file '{zip_file_path}' was not found in the current directory.")
else:
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(extract_dir)
        print(f"Successfully extracted '{zip_file_path}' to '{extract_dir}'.")
    except zipfile.BadZipFile:
        print(f"Error: The file '{zip_file_path}' is a corrupted zip file.")
    except Exception as e:
        print(f"An unexpected error occurred during extraction: {e}")

# After extraction, list the contents of the extracted directory to verify
if os.path.exists(extract_dir) and os.path.isdir(extract_dir):
    print(f"\nContents of '{extract_dir}':")
    print(os.listdir(extract_dir))
else:
    print(f"\nCould not verify contents of '{extract_dir}' as it does not exist or is not a directory.")
```

## Build the model

### Subtask:
Load the MobileNetV2 model, excluding the top classification layer, and add new layers for the classification task.


**Reasoning**:
Load the MobileNetV2 model excluding the top layer and add new layers for classification.



```python
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import AveragePooling2D, Dropout, Flatten, Dense
from tensorflow.keras.models import Model

# Load the MobileNetV2 base model
base_model = MobileNetV2(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))

# Freeze the base model
base_model.trainable = False

# Add new classification layers on top of the base model
x = base_model.output
x = AveragePooling2D(pool_size=(7, 7))(x)
x = Flatten(name="flatten")(x)
x = Dense(128, activation="relu")(x)
x = Dropout(0.5)(x)
# The number of units in the dense layer depends on the number of classes.
# Since the number of classes is unknown at this point, a placeholder is used.
# This needs to be updated with the actual number of classes later.
num_classes = 2 # Placeholder: Update with the actual number of classes
predictions = Dense(num_classes, activation="softmax")(x)

# Create the final model
model = Model(inputs=base_model.input, outputs=predictions)

model.summary()
```
